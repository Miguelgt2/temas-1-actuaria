```python
import math
import numpy as np

def exp(n, l):
    lista = []
    for a in range(n):
        u = np.random.random()  # Generate a uniform random number
        x = -(1/l) * np.log(1 - u)  # Exponential transformation
        lista.append(x)
    return np.array(lista)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rmarkovchain(lam,pi,m):
    P = np.array([[0.1,0.2,0.3,0.1,0.3],[0.2,0.1,0.2,0.2,0.3],[0.3,0.2,0.1,0.2,0.2],[0.1,0.2,0.2,0.1,0.4],[0.3,0.3,0.2,0.1,0.1]])
    states = [0,1,2,3,4]

    x=[np.random.choice(states, p = pi, size = 1)[0]] #simulamos x0 usando pi 
    x
    lista = []
    acum = []
    t = exp(1000,9).mean()
    t1 = t
    lista.append(t)
    acum.append(t)
    for i in range(1,m):
        
        estado_actual = x[-1] #Se ocupa la simulación anterior para revisar el siguiente
        t = exp(1000,9).mean()
        lista.append(t)
        t1=t+t1
        acum.append(t1)
        next_estado = np.random.choice(states, p = P[estado_actual])  #Realizamos la siguiente simulación
        
        x.append(next_estado)
        
    return np.array(x),np.array(lista),np.array(acum)
```
