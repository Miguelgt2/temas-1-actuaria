```python
def potencia(n,sim):
    lista=[]
    for i in range(sim):
        u=np.random.random()
        x=u**(1/n)
        lista.append(x)
    return np.array(lista)
```

```python
print(potencia(10,100).mean())
print(potencia(10,100).var())
```

```python
def rexp(lam,sim):
    lista=[]
    for i in range(sim):
        u=np.random.random()
        x=-1/lam*np.log(u)
        lista.append(x)
    return np.array(lista)
```

```python
rexp(1,100).mean()
```

```python
#simular gamma(n,lam)
rexp(1,10).sum()
```

```python
#simular gamma(n,lam) ideas bajas mias 
def mrexp(lam,sim):
    lista=[]
    x=0
    for i in range(sim):
        u=np.random.random()
        x1=-1/lam*np.log(u)
        x=x+x1
        lista.append(x)
    return np.array(lista)
```

```python
mrexp(1,10)
```

```python
#simular gamma(n,lam) profe
def rgamma(n,lam,sim):
    lista=[]
    for i in range(sim):
        y=rexp(lam,n).sum()
        lista.append(y)
    return np.array(lista)
```

```python
rgamma(10,1,10)
```

```python
def rgamma2(n,lam,sim):
    lista=[]
    for i in range(sim):
        u=np.random.random(n)
        x=-1/lam*np.log(u.prod())
        lista.append(x)
        return np.array(lista)
```

```python
rgamma2(10,1,1000).mean()
```
