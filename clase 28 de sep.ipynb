```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parametros
t=1.0 #timepo final / horizonte de tiempo
n=1000 #número de PArticiones
a=0.01
b=0.08
sigma=0.05
x_0=0.004
#PAso 1 Genrar una particion P
P=np.linspace(0,t,n+1)
#paso 2: Transforacion para optenrr Q
Q=np.log(P+1)/(2*a)
#paso 3: Generación el movimiento Browniano
dB=np.random.normal(0,np.sqrt(np.diff(np.exp(2*a*Q))),size=n)
#paso 4: calculo del proceso 
B=np.concatenate(([0],np.cumsum(dB)))
x_q=x_0*np.exp(-a*Q)+b*(1-np.exp(-a*Q))+(sigma-np.exp(-a*Q)/np.sqrt(1*a)*B)
```

```python
#grafica DEl Procesp de Ornstein-Uhlenbeck
plt.plot(P,x_q)
```

```python
# MEtodo de Euler 
#Parametros
alpha=0.05
sigma=2.5
b=0
x_0=0
t=1
n=1000
dt=t/n
# paso 1
Y=np.zeros(n+1)
Y[0]=x_0
time=np.linspace(0,t,n+1)
#paso 2: simular mediante Euler
for i in range(n):
    dw=np.random.normal(0,np.sqrt(dt))
    Y[i+1]=Y[i] + alpha*(b-Y[i])*dt + sigma*dw
    
```

```python
plt.plot(time,Y)
```
