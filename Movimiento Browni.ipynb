# movimiento browniano


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# valores de entrada
T=1 #horizonte de tiempo
N=10000 #Núnero de pasos
dt= T/N #tamaño de paso
t= np.linspace(0,T,N) #arreglo de tiempo partido en N pasos
#simular el proceso de Wiener (Movimiento Browniano)
B=np.zeros(N)
dB= np.random.normal(0,np.sqrt(dt),N)
np.cumsum(dB)
B=np.cumsum(dB)
plt.plot(t,B)
```

```python
#movimiento Browniano Geometrico
#parametros del MBg
S0=100 #precio inicial
mu=0.1 #media
sigma= 0.2 #desviacion estandar
T=1 #tiempo total (1 Año)
N=1000
dt= T/N

#iniciar simulacion
t=np.linspace(0,T,N+1)
S=np.zeros(N+1)
S[0]=S0 #el precio inicial es S0

#simular los incremenetos 
for i in range(1,N+1):
    z=np.random.normal(0,np.sqrt(dt))
    #z=np.random.normal(0,1)# ejemplo profe
    S[i]=S[i-1]*np.exp((mu-0.5*sigma**2)*dt+sigma*np.sqrt(dt)*z)
```

```python
plt.plot(t,S)
```
