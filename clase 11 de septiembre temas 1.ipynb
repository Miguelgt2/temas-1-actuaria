```python
import numpy as np
```

```python
def rnormalstd(n):
    lista=[]
    for _ in range(n):

        u1 = np.random.random()
        u2 = np.random.random()
        y1=-np.log(u1)
        y2=-np.log(u2)
        while(y2<((y1-1)**2)/2):
            u1 = np.random.random()
            u2 = np.random.random()
            y1=-np.log(u1)
            y2=-np.log(u2)
        if np.random.random()<0.5:
            lista.append(y1)
        else:
            lista.append(-y1)
    return np.array(lista)
```

```python
import matplotlib.pyplot as plt
plt.hist(rnormalstd(100000), bins=50)
plt.show()
```

```python
def rnorm(n,media,sigma):
    return sigma*rnormalstd(n)+media
    
```

```python
import matplotlib.pyplot as plt
plt.hist(rnorm(1000,10,1), bins=50)
plt.show()
```
